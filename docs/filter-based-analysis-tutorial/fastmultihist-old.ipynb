```python
%%cython --annotate
import cython
import numpy as np
cimport numpy as np
# These imports have to be in the same cell as the function and must be prefaced
# with %%cython

@cython.boundscheck(False)
@cython.wraparound(False)
def add2(double[:] x):
    cdef unsigned long i, N = x.size
    for i in range(N):
        x[i] += 1
```
